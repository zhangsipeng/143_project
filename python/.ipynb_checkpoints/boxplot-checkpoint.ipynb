```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = '/content/drive/MyDrive/Country Development Indicators.csv'


df = pd.read_csv(file_path)

# GDP
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='GDP')
plt.title('Boxplot of GDP')
plt.ylabel('GDP (in billions)')
plt.xlabel('Countries')
plt.grid(True)
plt.show()

# F rate
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='Fertility rate, total (births per woman)')
plt.title('Boxplot of Fertility Rate')
plt.ylabel('Fertility Rate (births per woman)')
plt.xlabel('Countries')
plt.grid(True)
plt.show()

```
